```python
import pandas as pd
payments = pd.read_csv('../data/payments.csv')

print(payments.columns)
```

```python
order_total = payments.groupby('order_id').agg(
    total_payment=('payment_value', 'sum'),
    installments=('installments', 'first')
).reset_index()
result = order_total.groupby('installments')['total_payment'].mean()
result.idxmax()
```
