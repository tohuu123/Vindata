```python
import pandas as pd

customers = pd.read_csv('../data/customers.csv')
orders = pd.read_csv('../data/orders.csv')

customers
# orders
```

```python
df = customers.merge(orders, on='customer_id')
df
```

```python
df = df[df['age_group'].notna()]

result = df.groupby('age_group').agg(
    total_orders=('order_id', 'count'),
    total_customers=('customer_id', 'nunique')
)
```

```python
result['avg_orders'] = result['total_orders'] / result['total_customers']
result['avg_orders'].idxmax()
```
